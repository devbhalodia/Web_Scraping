```python
table = soup.find_all("article", class_="product_pod")
'''
as you can clearly see in the above print statement, you can see that as we iterated on the table, it is giving us a collection of articles.
hence, we can say that when iterating on the table, we get i as each article. now, from this article, we need book title, book price, availability and
star rating.
'''
tit = []
for i in table:
    h3_tag = i.find("h3")
    if h3_tag:
        tit.append(h3_tag.text.strip())

print(tit)
```

```python
book_price = []
for i in table:
  p_tag = i.find("p", class_="price_color")
  if p_tag:
    book_price.append(p_tag.text.strip())

print(book_price)
```

```python
star_rating = []
for i in table:
  p_tag = i.find("p", class_="star-rating")
  if p_tag:
    star_rating.append(p_tag.get("class")[1])

print(star_rating)

'''
when we get class of any tag, we get split string. for eg: in the above p tag, class was : star-rating three.
so split:-  ["star-rating", "three"]
we simply accessed the index position 1. that's it.
this is done when we don't have any text but the most crucial information lies in the class of the tag.
'''
```

```python
availability = []
for i in table:
  p_tag = i.find("p", class_="instock availability")
  if p_tag:
    availability.append(p_tag.text.strip())

print(availability)
```

```python
'''
scraping the data from all 50 pages.
'''

url = "https://books.toscrape.com/catalogue/page-{}.html"

title = []
price = []
stock_availability = []
rating = []

for i in range(1,51):
  r = requests.get(url.format(i))
  soup = BeautifulSoup(r.text,"lxml")
  table = soup.find_all("article", class_="product_pod")
  for i in table:
      h3_tag = i.find("h3")
      if h3_tag:
          title.append(h3_tag.text.strip())

  print(title)


  for i in table:
    p_tag = i.find("p", class_="price_color")
    if p_tag:
      price.append(p_tag.text.strip())

  print(price)

  for i in table:
    p_tag = i.find("p", class_="star-rating")
    if p_tag:
      rating.append(p_tag.get("class")[1])

  print(rating)

  for i in table:
    p_tag = i.find("p", class_="instock availability")
    if p_tag:
      stock_availability.append(p_tag.text.strip())

  print(stock_availability)

```

```python
print(title)
print(price)
print(rating)
print(stock_availability)
```

```python
listoflist =[[]]
listoflist = list(zip(title, price, rating, stock_availability))

df = pd.DataFrame(listoflist, columns=["Title", "Price", "Rating", "Availability"])

print(df)
```

```python
df.to_csv('online_book_store.csv', index=False)
```
